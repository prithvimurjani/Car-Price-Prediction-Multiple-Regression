We now see a linear relationship with log price and others. Time to drop the price column from our data

```python
data_cleaned = data_cleaned.drop(['Price'],axis=1)
data_cleaned
```

Multicollinearity - it is possible that year and mileage are related

```python
data_cleaned.columns.values
```

Since sklearn cannot check this, we use statsmodels

Best way to check for multicollinearty is VIF method (Variance Inflation Factor)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
```

```python
#VIF = 1 ; NO MULTICOLLINEARITY
#VIF FROM 1 TO 5 IS OK
#VIF > 10 IS UNACCEPTABLE
```

```python
vif
```

```python
# We see that year is too correlated with the other variables so we remove it
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```

```python
data_no_multicollinearity
```

Creating dummy variables for categorical data

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity,drop_first = True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price','Mileage', 'EngineV',  'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

Building the regression

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state=365)
```

Creating the linear regression

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#To obtain predicted values
```

```python
yhat = reg.predict(x_train)
```

```python
plt.scatter(y_train,yhat)
plt.xlabel('Targets(y_train)',size=18)
plt.ylabel('Predictions(yhat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
#We must get a straight line of 45 degrees which tells us that for each training value, we get our expected y value
```

Plotting a residual graph = differences between target and predictions

```python
sns.distplot(y_train - yhat)
plt.title('Residual PDF',size=18)
```

```python
# We notice that the mean=0 and the graph is normally distributed which is what we wanted, however the graph has
# some values towards the negative side
```

```python
# This indicates that in some cases the inputs overestimate our targets
```

Calculating the R sq

```python
reg.score(x_train,y_train)
```

```python
# It is good result but could be better
```

Finding the weights and bias

```python
reg_summary = pd.DataFrame(inputs.columns.values,columns = ['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
# These are not easily interpretable however if a weight is positive, it indicates that the increase in the feature
# leads to an increase in price and log_price
```

Testing

```python
yhat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,yhat_test)
plt.xlabel('Targets(y_test)',size=18)
plt.ylabel('Predictions(yhat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
#We must get a straight line of 45 degrees which tells us that for each training value, we get our expected y value
```

```python
plt.scatter(y_test,yhat_test,alpha=0.5)
plt.xlabel('Targets(y_test)',size=18)
plt.ylabel('Predictions(yhat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
#We must get a straight line of 45 degrees which tells us that for each training value, we get our expected y value
```

```python
# Understanding the performance
df_pf = pd.DataFrame(np.exp(yhat_test),columns = ['Predictions'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test = y_test.reset_index(drop = True)
y_test.head()
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Predictions']
df_pf.head()
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
```
